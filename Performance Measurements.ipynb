```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

1. Test 1 : Raw compute capability - how does the throughput vary using different strategies?

```python
metric1 = pd.read_csv("GFlopsvN.csv", delimiter=',')
x = metric1["Size(N)"]
y1 = metric1["single-precision"]
y2 = metric1["double-precision"]
y3 = metric1["cpu"]
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("Number of Particles (N)")
plt.ylabel("GFlops")
plt.legend(['GPU-FP32', 'GPU-FP64', 'CPU'], loc='lower right')
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
```

2. Test 2 : Time taken to complete 10 timesteps of the simulations

```python
metric2 = pd.read_csv("TimevN.csv", delimiter=',')
x = metric2["Size(N)"]
y1 = metric2["time"]
y2 = metric2["cputime"]
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("Number of Particles (N)")
plt.ylabel("GFlops")
plt.legend(['Time for 10 steps (ms) - GPU', 'Time for 10 steps (ms) - CPU'], loc='upper left')
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
```

3. Test 2 : Effect of Tile size on throughput

```python
metric3 = pd.read_csv("BlockSizevN.csv", delimiter=',')
x = metric3["Size(N)"]
y1 = metric3["Gflops1"]
y2 = metric3["Gflops2"]
y3 = metric3["Gflops3"]
y4 = metric3["Gflops5"]
y5 = metric3["Gflops4"]
y6 = metric3["Gflops6"]
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.xlabel("Number of Particles (N)")
plt.ylabel("GFlops")
plt.legend(['blockSize = 64', 'blockSize = 128', 'blockSize = 256', 'blockSize = 512', 'blockSize = 1024', 'blockSize = 4096'], loc='lower right')
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
```
